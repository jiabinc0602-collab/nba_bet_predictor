```python
import pandas as pd
oddsdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\kaggle_odds_combined.csv', low_memory=False)
playersdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\nba_player_data.csv', low_memory=False)
teamsdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\nba_team_data.csv',low_memory=False)
oddsdf = oddsdf.drop_duplicates()
playersdf = playersdf.drop_duplicates()
teamsdf = teamsdf.drop_duplicates()
```

```python
#Dropping dates beyond the betting lines cutoff as they are irrelevant, sorted by time as well
cutoff_date = '2022-12-11'
playersdf['GAME_DATE'] = pd.to_datetime(playersdf['GAME_DATE'])
playersdf = playersdf.loc[(playersdf['GAME_DATE'] < cutoff_date)]
playersdf = playersdf.sort_values(by='GAME_DATE')
playersdf = playersdf.reset_index(drop=True)
```

```python
#Dropping dates beyond the betting lines cutoff as they are irrelevant, sorted by time as well
teamsdf['GAME_DATE'] = pd.to_datetime(teamsdf['GAME_DATE'])
teamsdf = teamsdf.loc[(teamsdf['GAME_DATE'] < cutoff_date)]
teamsdf = teamsdf.sort_values(by='GAME_DATE')
teamsdf = teamsdf.reset_index(drop=True)
```

```python
columnsToFetch = [
    'game_date',
    'NBA_team_name',
    'home_away',
    'close_moneyline_line',
    'close_spread_line',
    'close_total_line'
]
#Filter by columns of data I am going to use
oddsdf = oddsdf[columnsToFetch]

oddsdf = oddsdf.dropna(subset=['close_spread_line', 'close_moneyline_line'])

oddsdf = oddsdf.reset_index(drop=True)

print(f"Total Rows: {len(oddsdf)}")
oddsdf.head()
```

```python
nba_teams = [
    "Atlanta Hawks",
    "Boston Celtics",
    "Brooklyn Nets",
    "Charlotte Hornets",
    "Chicago Bulls",
    "Cleveland Cavaliers",
    "Dallas Mavericks",
    "Denver Nuggets",
    "Detroit Pistons",
    "Golden State Warriors",
    "Houston Rockets",
    "Indiana Pacers",
    "LA Clippers",
    "Los Angeles Lakers",
    "Memphis Grizzlies",
    "Miami Heat",
    "Milwaukee Bucks",
    "Minnesota Timberwolves",
    "New Orleans Pelicans",
    "New York Knicks",
    "Oklahoma City Thunder",
    "Orlando Magic",
    "Philadelphia 76ers",
    "Phoenix Suns",
    "Portland Trail Blazers",
    "Sacramento Kings",
    "San Antonio Spurs",
    "Toronto Raptors",
    "Utah Jazz",
    "Washington Wizards"
]
teamsdf = teamsdf[teamsdf['TEAM_NAME'].isin(nba_teams)]
playersdf = playersdf[playersdf['TEAM_NAME'].isin(nba_teams)]
print(f"Total Rows:{len(teamsdf)}")
teamsdf.head(10)
```

```python
#Calculating a players plus minus rolling average to find out their recent "team impact" to measure their impact
playersdf['PLUS_MINUS_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['PLUS_MINUS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
playersdf
```

```python
#Calculate rolling averages for teams to identify matchup and recent performance
teamsdf['PTS_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['PTS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['PLUS_MINUS_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['PLUS_MINUS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['OREB_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['OREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['DREB_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['DREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['TOV_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['TOV'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['STL_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['STL'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf
```

```python
oddsdf['game_date'] = pd.to_datetime(oddsdf['game_date'])
final_df = pd.merge(
    teamsdf,
    oddsdf,
    left_on=['TEAM_NAME', 'GAME_DATE'],
    right_on=['NBA_team_name', 'game_date'],
    how='left'
)
final_df = final_df.drop(['game_date', 'NBA_team_name'], axis = 1)

final_df
```

```python
#Added column "Covered" to find out if the win margin covered the betting spread
model_df = final_df.dropna(subset=['close_spread_line']).copy()

model_df['covered'] = ((model_df['PLUS_MINUS'] + model_df['close_spread_line'])>0).astype(int)
model_df = model_df.reset_index()
model_df
```

```python
#Separating dataframes from training and testing, (Testing dataframe contains the betting lines)
split_date = '2021-09-30'
train_df = model_df[model_df['GAME_DATE'] < split_date].copy()
test_df = model_df[model_df['GAME_DATE']>=split_date].copy()
print(f"Training on {len(train_df)} games")
print(f"Testing on {len(test_df)} games")
```
