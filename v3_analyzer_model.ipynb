```python
import pandas as pd
oddsdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\kaggle_odds_combined.csv', low_memory=False)
playersdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\nba_player_data.csv', low_memory=False)
teamsdf = pd.read_csv(r'C:\Users\jiabi\NBA_Analyzerv2\nba_team_data.csv',low_memory=False)
oddsdf = oddsdf.drop_duplicates()
playersdf = playersdf.drop_duplicates()
teamsdf = teamsdf.drop_duplicates()
```

```python
#Dropping dates beyond the betting lines cutoff as they are irrelevant, sorted by time as well
cutoff_date = '2022-12-11'
playersdf['GAME_DATE'] = pd.to_datetime(playersdf['GAME_DATE'])
playersdf = playersdf.loc[(playersdf['GAME_DATE'] < cutoff_date)]
playersdf = playersdf.sort_values(by='GAME_DATE')
playersdf = playersdf.reset_index(drop=True)

#Dropping dates beyond the betting lines cutoff as they are irrelevant, sorted by time as well
teamsdf['GAME_DATE'] = pd.to_datetime(teamsdf['GAME_DATE'])
teamsdf = teamsdf.loc[(teamsdf['GAME_DATE'] < cutoff_date)]
teamsdf = teamsdf.sort_values(by='GAME_DATE')
teamsdf = teamsdf.reset_index(drop=True)
```

```python
columnsToFetch = [
    'game_date',
    'NBA_team_name',
    'home_away',
    'close_moneyline_line',
    'close_spread_line',
    'close_total_line'
]
#Filter by columns of data I am going to use
oddsdf = oddsdf[columnsToFetch]

oddsdf = oddsdf.dropna(subset=['close_spread_line', 'close_moneyline_line'])

oddsdf = oddsdf.reset_index(drop=True)

print(f"Total Rows: {len(oddsdf)}")
oddsdf.head()
```

```python
nba_teams = [
    "Atlanta Hawks",
    "Boston Celtics",
    "Brooklyn Nets",
    "Charlotte Hornets",
    "Chicago Bulls",
    "Cleveland Cavaliers",
    "Dallas Mavericks",
    "Denver Nuggets",
    "Detroit Pistons",
    "Golden State Warriors",
    "Houston Rockets",
    "Indiana Pacers",
    "LA Clippers",
    "Los Angeles Lakers",
    "Memphis Grizzlies",
    "Miami Heat",
    "Milwaukee Bucks",
    "Minnesota Timberwolves",
    "New Orleans Pelicans",
    "New York Knicks",
    "Oklahoma City Thunder",
    "Orlando Magic",
    "Philadelphia 76ers",
    "Phoenix Suns",
    "Portland Trail Blazers",
    "Sacramento Kings",
    "San Antonio Spurs",
    "Toronto Raptors",
    "Utah Jazz",
    "Washington Wizards"
]
teamsdf = teamsdf[teamsdf['TEAM_NAME'].isin(nba_teams)]
playersdf = playersdf[playersdf['TEAM_NAME'].isin(nba_teams)]
print(f"Total Rows:{len(teamsdf)}")
teamsdf.head(10)
```

```python
#Calculating a players plus minus rolling average to find out their recent "team impact" to measure their impact
playersdf['PLUS_MINUS_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['PLUS_MINUS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
playersdf['OREB_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['OREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
playersdf['DREB_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['DREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
playersdf['TOV_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['TOV'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
playersdf['STL_ROLLING'] = playersdf.groupby(['PLAYER_ID', 'SEASON_ID'])['STL'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))

#Calculating rolling averages for teams
teamsdf['PTS_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['PTS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['PLUS_MINUS_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['PLUS_MINUS'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['OREB_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['OREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['DREB_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['DREB'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['TOV_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['TOV'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
teamsdf['STL_ROLLING'] = teamsdf.groupby(['TEAM_ID', 'SEASON_ID'])['STL'].transform(lambda x : x.rolling(10, min_periods = 1).mean().shift(1))
```

```python
oddsdf['game_date'] = pd.to_datetime(oddsdf['game_date'])
final_df = pd.merge(
    teamsdf,
    oddsdf,
    left_on=['TEAM_NAME', 'GAME_DATE'],
    right_on=['NBA_team_name', 'game_date'],
    how='left'
)
final_df = final_df.drop(['game_date', 'NBA_team_name'], axis = 1)

final_df
```

```python
#Finds the players that are going to be playing in the next game and calculates the roster statistics based on active players
active_players = playersdf[playersdf['MIN'] > 0].copy()
active_roster = active_players.groupby(['GAME_ID', 'TEAM_ID']).agg({
    'PLUS_MINUS_ROLLING' : 'sum',
    'OREB_ROLLING': 'sum',
    'DREB_ROLLING': 'sum',
    'TOV_ROLLING': 'sum',
    'STL_ROLLING': 'sum',
    'MIN': 'sum'
}).reset_index()

active_roster = active_roster.rename(columns={
    'PLUS_MINUS_ROLLING': 'ROSTER_PLUS_MINUS',
    'OREB_ROLLING': 'ROSTER_OREB',
    'DREB_ROLLING': 'ROSTER_DREB',
    'TOV_ROLLING': 'ROSTER_TOV',
    'STL_ROLLING': 'ROSTER_STL',
    'MIN': 'ROSTER_MIN_TOTAL'
})

final_df = final_df.merge(
    active_roster,
    on=['GAME_ID','TEAM_ID'],
    how='left'
)
#Finding out if a team is playing back to back games
final_df = final_df.sort_values(by=['TEAM_ID', 'GAME_DATE'])
final_df['REST_DAYS'] = final_df.groupby('TEAM_ID')['GAME_DATE'].diff().dt.days
final_df['BACK_TO_BACK'] = (final_df['REST_DAYS'] == 1).astype(int)

#Added column "Covered" to find out if the win margin covered the betting spread
model_df = final_df.dropna(subset=['close_spread_line']).copy()

#For a betting line to be covered, it the plus minus win margin must be greater than the spread line
model_df['covered'] = ((model_df['PLUS_MINUS'] + model_df['close_spread_line'])>0).astype(int)
model_df = model_df.reset_index()

#Creating opponent columns to analyze matchup
model_df['OPP_ABBREVIATION'] = model_df['MATCHUP'].apply(lambda x: x.split(' ')[-1])

columns_to_copy = [
    'GAME_DATE',
    'TEAM_ABBREVIATION',
    'PTS_ROLLING','PLUS_MINUS_ROLLING','OREB_ROLLING','DREB_ROLLING','TOV_ROLLING','STL_ROLLING',
    'ROSTER_PLUS_MINUS','ROSTER_OREB','ROSTER_DREB','ROSTER_TOV','ROSTER_STL'
]

lookup_df = model_df[columns_to_copy].copy()

columns_rename = {
    'TEAM_ABBREVIATION' : 'OPP_ABBREVIATION',
    'PTS_ROLLING' : 'OPP_PTS_ROLLING',
    'PLUS_MINUS_ROLLING' : 'OPP_PLUS_MINUS',
    'OREB_ROLLING' : 'OPP_OREB_ROLLING',
    'DREB_ROLLING' : 'OPP_DREB_ROLLING',
    'TOV_ROLLING' : 'OPP_TOV_ROLLING',
    'STL_ROLLING' : 'OPP_STL_ROLLING',
    #Roster renames
    'ROSTER_PLUS_MINUS': 'OPP_ROSTER_PLUS_MINUS',
    'ROSTER_OREB': 'OPP_ROSTER_OREB',
    'ROSTER_DREB': 'OPP_ROSTER_DREB',
    'ROSTER_TOV': 'OPP_ROSTER_TOV',
    'ROSTER_STL': 'OPP_ROSTER_STL'
}

lookup_df = lookup_df.rename(columns=columns_rename)

model_df = pd.merge(
    model_df,
    lookup_df,
    on = ['GAME_DATE','OPP_ABBREVIATION'],
    how = 'left'
)

#Adding columns for matchup analysis
model_df['ROSTER_OREB_ADVANTAGE'] = model_df['ROSTER_OREB'] - model_df['OPP_ROSTER_DREB']
model_df['ROSTER_DREB_ADVANTAGE'] = model_df['ROSTER_DREB'] - model_df['OPP_ROSTER_OREB']

model_df['ROSTER_GOOD_CHAOS'] = model_df['ROSTER_STL'] + model_df['OPP_ROSTER_TOV']
model_df['ROSTER_BAD_CHAOS'] = model_df['ROSTER_TOV'] + model_df['OPP_ROSTER_STL']

model_df['ROSTER_PLUS_MINUS_ADVANTAGE'] = model_df['ROSTER_PLUS_MINUS'] - model_df['OPP_ROSTER_PLUS_MINUS']
model_df.columns.to_list()
```

```python
#Mapping home to numbers for ML algorithm to process
model_df['is_home'] = model_df['home_away'].map({'home' : 1, 'away' : 0})
```

```python
#Separating dataframes from training and testing, (Testing dataframe contains the betting lines)
split_date = '2021-09-30'
train_df = model_df[model_df['GAME_DATE'] < split_date].copy()
test_df = model_df[model_df['GAME_DATE']>=split_date].copy()
print(f"Training on {len(train_df)} games")
print(f"Testing on {len(test_df)} games")
train_df.columns
```

```python
#Columns to analyze
correlation_columns = [
    'covered',
    'close_spread_line',
    'ROSTER_PLUS_MINUS_ADVANTAGE',
    'ROSTER_OREB_ADVANTAGE',
    'ROSTER_DREB_ADVANTAGE',
    'ROSTER_BAD_CHAOS',
    'ROSTER_GOOD_CHAOS',
    'BACK_TO_BACK',
    'is_home'
]

correlation_matrix = train_df[correlation_columns].corr()

#Finding which statistics has the strongest correlation with a betting spread being "covered"
print(correlation_matrix['covered'].sort_values(ascending = False))
```

```python
%pip install scikit-learn
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score

rf = RandomForestClassifier(n_estimators = 100, min_samples_split=10, random_state = 1)

predictors = [
    'close_spread_line',
    'is_home',
    'ROSTER_PLUS_MINUS_ADVANTAGE',
    'ROSTER_OREB_ADVANTAGE',
    'ROSTER_DREB_ADVANTAGE',
    'ROSTER_BAD_CHAOS',
    'ROSTER_GOOD_CHAOS',
    'BACK_TO_BACK'
]

rf.fit(train_df[predictors], train_df['covered'])

preds = rf.predict(test_df[predictors])
```

```python
acc = accuracy_score(test_df['covered'],preds)
print(f"Model Accuracy: {acc}")
```

```python
importances = rf.feature_importances_
feature_importances_df = pd.DataFrame({
    'feature' : predictors,
    'importance' : importances
    }).sort_values('importance', ascending = False)

print(feature_importances_df)
```

```python
probs = rf.predict_proba(test_df[predictors])[:, 1]

results_df = pd.DataFrame({
    'actual': test_df['covered'],
    'prediction': preds,
    'probability': probs
})

high_confidence_bets = results_df[results_df['probability'] > 0.67]

print(f"Total High Confidence Bets: {len(high_confidence_bets)}")
print(f"Accuracy on High Conf Bets: {accuracy_score(high_confidence_bets['actual'], high_confidence_bets['prediction'])}")

low_confidence_bets = results_df[results_df['probability'] <= 0.65]
print(f"Total Low Confidence Bets: {len(low_confidence_bets)}")
print(f"Accuracy on Low Conf Bets: {accuracy_score(low_confidence_bets['actual'], low_confidence_bets['prediction'])}")
```

```python
#Calculating the profit for using this model, assuming that average bet is around -110
confident_bets = results_df[results_df['probability'] > 0.60].copy()
confident_bets.head()
def calculate_profit(row):
    if row['prediction'] == row['actual']:
        return 1.00
    else:
        return -1.10

confident_bets['profit'] = confident_bets.apply(calculate_profit, axis=1)

total_profit = confident_bets['profit'].sum()
ROI = total_profit/(len(confident_bets)*1.1)*100

#Total profit is in units, ROI is in %
print(f"Total bets: {len(confident_bets)}")
print(f"Total Profit: {total_profit}")
print(f"Return On Investment: {ROI}")
```
